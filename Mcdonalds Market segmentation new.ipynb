```python
#Label encoding for categorical - Converting 11 cols with yes/no

from sklearn.preprocessing import LabelEncoder

def labelling(x):
    df[x] = LabelEncoder().fit_transform(df[x])
    return df

cat = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']

for i in cat:
    labelling(i)

df
```

```python
#Considering only first 11 attributes
df_eleven = df.loc[:,cat]
df_eleven
```

```python
#Considering only the 11 cols and converting it into array
x = df.loc[:,cat].values
x
```

```python
#Principal component analysis

from sklearn.decomposition import PCA
from sklearn import preprocessing

pca_data = preprocessing.scale(x)

pca = PCA(n_components=11)
pc = pca.fit_transform(x)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11']
pf = pd.DataFrame(data = pc, columns = names)
pf
```

```python
#Explained Variance Ratio (from PC1 to PC11)

pca.explained_variance_ratio_
```

```python
#cumulative explained variance

np.cumsum(pca.explained_variance_ratio_)

#The first few principal components capture most of the meaningful patterns in data.
#around PC8 or PC9 covering ~90-93% variance in data.
```

```python
# correlation coefficient between original variables and the component

loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame(loadings.T, columns=pc_list)
loadings_df['variable'] = df_eleven.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

```python
#Correlation matrix plot for loadings 
plt.rcParams['figure.figsize'] = (20,15)
ax = sns.heatmap(loadings_df, annot=True, cmap='Spectral')
plt.show()
```

```python
#Scree plot (Elbow test)- PCA

from bioinfokit.visuz import cluster

cluster.screeplot(obj=[pc_list,
                       pca.explained_variance_ratio_],
                       show=True,dim=(10,5))

#we can select columns till 8 or 9 and remove the remaining columns to reduce dimension 
```

```python
# get PC scores
pca_scores = PCA().fit_transform(x)

# get 2D biplot
cluster.biplot(cscore=pca_scores, 
               loadings=loadings, 
               labels=df.columns.values, 
               var1=round(pca.explained_variance_ratio_[0]*100, 2),
               var2=round(pca.explained_variance_ratio_[1]*100, 2),
               show=True,dim=(10,5))

# Direction of Loadings : 
# Features pointing in similar directions are positively correlated (people who think fattening, greasy. think MD is also disgusting) 
# opposite directions suggest negative correlation.
```

# EXTRACTING SEGMENTS

```python
#Using k-means clustering analysis

from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()

visualizer = KElbowVisualizer(model, k=(1,12)).fit(df_eleven)
visualizer.show()

#we can make 4 k-means clusters 
```

```python
#K-means clustering 

kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(df_eleven)

df['cluster_num'] = kmeans.labels_ #adding to df

print (kmeans.labels_) #Label assigned for each data point
print (kmeans.inertia_) #gives within-cluster sum of squares. 
print(kmeans.n_iter_) #number of iterations that k-means algorithm runs to get a minimum within-cluster sum of squares
print(kmeans.cluster_centers_) #Location of the centroids on each cluster. 
```

```python
#To see each cluster size

from collections import Counter
Counter(kmeans.labels_)
```

```python
#Visulazing clusters

sns.scatterplot(data=pf, x="pc1", y="pc2", hue=kmeans.labels_,palette='tab10')

plt.scatter(kmeans.cluster_centers_[:,0], 
            kmeans.cluster_centers_[:,1], 
            marker="X", c="r", s=80, 
            label="centroids")

plt.legend()
plt.show()
```

# Selecting target segment

```python
##Calculating the mean

#Visit frequency

df['VisitFrequency'] = LabelEncoder().fit_transform(df['VisitFrequency'])

visit = df.groupby('cluster_num')['VisitFrequency'].mean()
visit = visit.to_frame().reset_index()
visit
```

```python
#Like

df['Like'] = LabelEncoder().fit_transform(df['Like'])

Like = df.groupby('cluster_num')['Like'].mean()
Like = Like.to_frame().reset_index()
Like
```

```python
#Gender

df['Gender'] = LabelEncoder().fit_transform(df['Gender'])

Gender = df.groupby('cluster_num')['Gender'].mean()
Gender = Gender.to_frame().reset_index()
Gender
```

```python
segment = Gender.merge(Like, on='cluster_num', how='left').merge(visit, on='cluster_num', how='left')
segment

#observations:
#cluster 3 has good liker score and 2 highest frequency
```

```python
#Target segments (comparing VisitFrequency and Like scores)

plt.figure(figsize = (9,4))
sns.scatterplot(x = "VisitFrequency", y = "Like",data=segment,s=400, color="r")
plt.title("Simple segment evaluation plot for the Mcdonalds data set",
          fontsize = 15) 
plt.xlabel("Visit", fontsize = 12) 
plt.ylabel("Like", fontsize = 12) 
plt.show()
```

# hierarchical clustering

```python
df1=pd.read_csv('mcdonalds.csv')

df1['Like']= df1['Like'].replace({'I hate it!-5': '-5','I love it!+5':'+5'})



def labelling(x):
    df1[x] = LabelEncoder().fit_transform(df1[x])
    return df1

cat = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']

for i in cat:
    labelling(i)

df2=df1.copy()
df1 = df1.drop(columns=['Gender', 'VisitFrequency'])
df1
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df1 = scaler.fit_transform(df1)

linkage_matrix = linkage(df1, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix)

cutoff = 51
plt.axhline(y=cutoff, color='red', linestyle='--')

plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
num_clusters = 4  # Choosing the number of clusters based on the dendrogram cutoff
df1=pd.DataFrame(df1)
df1['Hierarchical_Clustering']= fcluster(linkage_matrix, num_clusters, criterion='maxclust')
df1.columns = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap', 'tasty', 
               'expensive', 'healthy', 'disgusting', 'Like', 'Age','Hierarchical_Clustering']
df1
```

```python
plt.figure(figsize=(9, 6))
sns.scatterplot(x=df2["VisitFrequency"], y=df2["Like"], hue=df1["Hierarchical_Clustering"], palette="Set2", s=100)
plt.title("Hierarchical Clustering Visualization")
plt.xlabel("Visit Frequency")
plt.ylabel("Like Score")
plt.legend(title="Clusters")
plt.show()
```

```python
cluster_means = df1.groupby("Hierarchical_Clustering").mean()

plt.figure(figsize=(10, 6))

sns.heatmap(cluster_means, cmap="coolwarm", annot=True, fmt=".2f")

plt.title("Hierarchical Clustering Feature Averages")
plt.xlabel("Features")
plt.ylabel("Clusters")
plt.show()
```
